# Certification project 1
This is a project consisting of all the lessons of the first part on Algebra with Python. This project consists of:


1.   Addition
2.   Subtraction
3.   Multiplication
4.   Division
5.   Test is a number is prime or composite
6.   Determine the prime factors of number
7.   Factor expressions
8.   Factor square roots
9.   Evaluate expressions
10.  Solving for x


```python
import math
import sympy
from sympy import *

# menu
print("Hello, welcome to my project collection, here are my available functions:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Prime or composite")
print("6. Prime factors")
print("7. Factor expressions")
print("8. Factor square roots")
print("9. Evaluate Expressions")
print("10. Solve for x")
choice = int(input("Enter your choice by typing the number: "))

def add():
    a = float(input('\nEnter a number: '))
    b = float(input('Enter another number: '))
    sum = a+b
    print(f"\n{a} + {b} = {sum}")

def subtract():
    a = float(input('\nEnter a number: '))
    b = float(input('Enter another number: '))
    diff = a-b
    print(f"\n{a} - {b} = {diff}")

def multiply():
    a = float(input('\nEnter a number: '))
    b = float(input('Enter another number: '))
    product = a*b
    print(f"\n{a} * {b} = {product}")

def divide():
    a = float(input('\nEnter a number: '))
    b = float(input('Enter another number: '))
    quotient = a/b
    print(f"\n{a} / {b} = {quotient}")

def prime_or_comp():
    number = int(input("\nEnter a positive integer: "))
    # make prime as initial value
    prime_or_comp = "prime"
    # loop through the number to check for factor
    for test_number in range(2,number):
        # test if the number is composite
        if number % test_number == 0:
            prime_or_comp = "composite"
    print(f"\nYour selected number is {prime_or_comp}")

def prime_factors():
    number = int(input('\nEnter an integer: '))
    pfactors = []  # store prime factors
    # loop through the number
    for test_factor in range(2, number + 1):
        # start from 2, since 1 is not prime
        if number % test_factor == 0:
            # check if test_factor is prime or composite
            prime_or_comp = "prime"
            for i in range(2, int(math.sqrt(test_factor)) + 1):
                if test_factor % i == 0:
                    prime_or_comp = "comp"
                    break
            if prime_or_comp == "prime":
                pfactors.append(test_factor)
    print(f"\nPrime factors of {number}: {pfactors}")

def factor_expressions():
    # initialize variables
    x, y, z = symbols('x y z')
    local_dict = {'x': x, 'y': y, 'z': z}
    # get user input
    original_eq = input('\nEnter an expression (in terms of x, y, or z) to factor: 0 = ')
    original_eq = original_eq.replace('^', '**')
    # parse input
    eq1 = sympify(original_eq, locals=local_dict)
    # factor expression
    factored_form = factor(eq1)
    # display result
    print("\nHere's the result:")
    display(Eq(eq1, factored_form))

def factor_square_roots():
    n = int(input('\nWithout the radical, enter a square root to factor: '))
    # determine upper limit for checking
    upper_limit = math.floor(math.sqrt(n)) + 1
    # initialize variables to be updated
    max_factor = 1
    other_factor = 1
    square_root = 1
    # check for square factors
    for maybe_factor in range(1, upper_limit):
        if n % (maybe_factor**2) == 0:
            max_factor = maybe_factor**2
    # update variables
    other_factor = n/max_factor
    square_root = int(math.sqrt(max_factor))
    other_factor = int(other_factor)
    output = square_root*sympy.sqrt(other_factor)
    print("\nThe factored form is: ")
    display(output)

def evaluate_expression():
    # initialize symbols
    x, y, z = symbols('x y z')
    # input equation
    expr = input("\nEnter an expression: ")
    expr = expr.replace('^', '**')
    # input value to be substituted
    val = float(input("Enter value to be substituted: "))
    result = sympify(expr).subs(x, val)
    print(f"\nYour expression is: ")
    display(sympify(expr))
    print(f"f({val}) = {float(result):.3f}") # output is in 3 d.p.

def solver():
    # initialize 'x' as a variable
    x = symbols('x')
    eq = input('\nEnter an equation to solve for x: 0 = ')
    # solve for x
    sol = solve(eq, x)
    print("\nYour equation")
    display(sympify(eq))
    print("Has the following solutions: ")
    for s in sol:
        display(Eq(x,s))

# check for user input choice
if choice == 1:
    add()
elif choice == 2:
    subtract()
elif choice == 3:
    multiply()
elif choice == 4:
    divide()
elif choice == 5:
    prime_or_comp()
elif choice == 6:
    prime_factors()
elif choice == 7:
    factor_expressions()
elif choice == 8:
    factor_square_roots()
elif choice == 9:
    evaluate_expression()
elif choice == 10:
    solver()
else:
    print("Your choice is Invalid")
```
